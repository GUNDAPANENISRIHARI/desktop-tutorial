```python
import csv
from datetime import datetime

# List to store customer contacts
contacts = []

# Function to add a contact
def add_contact(name, email, phone, notes, last_contact_date):
    contact = {
        'name': name,
        'email': email,
        'phone': phone,
        'notes': notes,
        'last_contact_date': last_contact_date
    }
    contacts.append(contact)
    print(f"Contact for {name} added successfully!")

# Function to update a contact
def update_contact(name, new_email=None, new_phone=None, new_notes=None, new_last_contact_date=None):
    for contact in contacts:
        if contact['name'] == name:
            if new_email:
                contact['email'] = new_email
            if new_phone:
                contact['phone'] = new_phone
            if new_notes:
                contact['notes'] = new_notes
            if new_last_contact_date:
                contact['last_contact_date'] = new_last_contact_date
                print(f"Contact for {name} updated successfully!")
            return
    print(f"Contact for {name} not found.")

# Function to delete a contact
def delete_contact(name):
    global contacts
    contacts = [contact for contact in contacts if contact['name'] != name]
    print(f"Contact for {name} deleted successfully!")

# Function to view all contacts
def view_contacts():
    if not contacts:
        print("No contacts available.")
        return
    for contact in contacts:
        print(f"Name: {contact['name']}, Email: {contact['email']}, Phone: {contact['phone']}, "
              f"Notes: {contact['notes']}, Last Contact: {contact['last_contact_date']}")

# Function to export contacts to CSV
def export_to_csv(filename='contacts.csv'):
    if not contacts:
        print("No contacts to export.")
        return
    keys = contacts[0].keys()
    with open(filename, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=keys)
        writer.writeheader()
        writer.writerows(contacts)
    print(f"Contacts exported to {filename}.")

# Example usage
add_contact("Hari", "hari123@gmail.com", "9219564845", "First meeting", str(datetime.now()))
add_contact("Surya", "surya246@gmail.com", "5572568528", "Doing job", str(datetime.now()))
add_contact("Divya", "divya@gmail.com", "9584235745", "Good morning", str(datetime.now()))
add_contact("Sravya", "sravya2000@gmail.com", "9632547164", "Happy birthday", str(datetime.now()))

# View contacts
view_contacts()

# Update contact
update_contact("Hari", new_notes="Second meeting", new_last_contact_date=str(datetime.now()))

# Delete a contact
delete_contact("Sravya")

# View contacts after update and delete
view_contacts()

# Export contacts to CSV
export_to_csv("customer_contacts.csv")
```
